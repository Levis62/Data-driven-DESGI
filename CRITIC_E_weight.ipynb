```python
df=data.iloc[:,1:]
```

```python
S=df.std() #计算标准差
S
```

```python
corr=df.corr() #计算相关系数
corr
```

```python
R=(1-corr).sum()
R
```

```python
C=S*R
C
```

```python
W=C/sum(C)
W
```

```python
E_weight_CRITIC = pd.DataFrame()
```

```python
E_weight_CRITIC=W
```
