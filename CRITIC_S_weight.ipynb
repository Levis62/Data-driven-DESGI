```python
data = pd.read_excel('C:\\Users\\student\\Desktop\\PythonLearning\\Data_driven ESG data\\CRITIC\\S data.xlsx')
data.head()
```

```python
df=data.iloc[:,1:]
S=df.std() #计算标准差
S
```

```python
corr=df.corr() #计算相关系数
corr
```

```python
R=(1-corr).sum()
R
```

```python
C=S*R
C
```

```python
W=C/sum(C)
W
```

```python
S_weight_CRITIC = pd.DataFrame()
S_weight_CRITIC=W
S_weight_CRITIC
S_weight_CRITIC.to_csv('C:\\Users\\student\\Desktop\\PythonLearning\\Data_driven ESG data\\CRITIC\\S_weight_CRITIC.csv')
```
