```python
def meboot(matrix, alpha=1):
    nrows, ncols = matrix.shape
    b = 1000 #number of bootstrap samples
    w = np.zeros(ncols)
    for i in range(b):
        resample = matrix[np.random.randint(nrows, size=nrows), :]
        p = resample / np.sum(resample, axis=0) #normalize to probabilities
        w += -np.sum(p * np.log(p), axis=0)
    w = w / b #average over bootstrap samples
    w += alpha * np.std(w) #add penalty term
    w = np.exp(-w) #convert to weights
    w = w / np.sum(w) #normalize to sum to 1
    return w
weights = meboot(data)
print("Input data:")
print(data)
print()
print("Weights:")
print(weights)
```
